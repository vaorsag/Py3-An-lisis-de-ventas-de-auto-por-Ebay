### Análisis de la columna Brand

La idea es explorar las variaciones que ha sufrido una mima marca. Para ello se utiliza la técnica de agregación. Los pasos siguientes para raelizar tal técnica son:

- Identificar los valores únicos que se desean agregar
- Crear un diccionario vacio en el cual se incorporara los datos agregados
- Utilizar un loop por cada "unique" valor y con el:
    - Filtrar la Data Frame por el unique value.
    - Calcular el valor promedio de la columna de interes
    - Asignar el valor promedio junto con su key al diccionario

```python
autos_top = autos["brand"].value_counts().head(5)  # se eligen las marcas top 20 para agregar
print(autos_top)
```

```python
sel_brands = autos_top.index.unique()                    # se extraen los nombres de las marcas
brands_mean = {}                                        # diccionario vacio a utilizar
print(sel_brands)
```

```python
for elem in sel_brands:
 selrows = autos[autos["brand"]== elem ]
 mean_price = selrows["price"].mean()
 brands_mean[elem] = int(mean_price)
print(brands_mean)
```

```python
print( "Diferencia de precion entre la mas popular y la quinta mas popular:",brands_mean['volkswagen'] - brands_mean['opel'])
print( "Diferencia de precion entre la mas popular y la mas cara:",brands_mean['audi'] - brands_mean['volkswagen'])
```

* Audi, BMW and Mercedes Benz las mas caras
* Opel la mas barata
* Volkswagen en el medio


### Análisis de la columna Kilometros recorridos vs Precios

Se quiere saber si existe alguna relacion entre el precio y los kilometros recoridos por los autos. Para ello se utilizara una combinacion de marca-precio-odometro, creando una sola columna para tales propósitos, es decir se combinarán dos series de nuestro Data Frame. Los métodos a utilizar son: 
pandas series constructor y pandas dataframe constructor.

En primer lugar se debe obtener el diccionario que falta: Kilometros diccionario:


```python
km_mean = {} 
for elem in sel_brands:
 selrows = autos[autos["brand"]== elem ]
 mean_km = selrows["odometer_km"].mean()
 km_mean[elem] = int(mean_km)
print(km_mean)
```

Con esto ya se tiene marca-precio y marca-kilometros diccionarios. Ahora se procede a transformacion de estos diccionarios un dataFrame de una columna utilizando los panda seires y panda dataframe constructors. 

```python
bmp_series = pd.Series(brands_mean) 
print(bmp_series) 
```

Como constructor (pandas series) crea un objeto de class pd.Series, donde transforma el diccionario en un objeto. La key en el diccionario se convierte en index y value en su valor en una columna, ahora falta darle nombre a la columna y para ello se utiliza el Data Frame constructor, el que crea un nuevo DataFrame de una columna.  

```python
df = pd.DataFrame(bmp_series, columns=['price_mean'])
df
```

```python
km_series = pd.Series(km_mean) 
print(km_series) 
```

Agrenando la serie de kilometros al dataframe antes creado:

```python
df["km_mean"] = km_series
df
```

Finalmente se observa que el kilometrage por cada marca no varia demasiado a diferencia del precio en cada marca,lo que se puede suponer que en cuanto a rendimiento tanto los autos mas caros como los mas baratos permiten realizar el mismo recorrido,De esta forma, finalmente se mostrara una tabla de cuanto sale cada kilometro que se recorre por cada auto ( solo en cuanto a su precio, claro esta)


```python
costo_km = df["km_mean"] / df["price_mean"]
print(costo_km)
```

Vemos que la marca Opel nos da mas de el doble de kilometros por cada euro que se paga por los mismos kilometros recorridos por los autos mas caros.
